Id and idhogar are unique identifiers.  Looking at the remaining 3 object columns:

```python
train['dependency'].unique()
```

Since we have a dependency squared column and a edjefe squared column, we can easily take the square root of that as a fix.

```python
train[['dependency', 'SQBdependency', 'edjefe', 'SQBedjefe']].head()
```

```python
train['dependency'] = np.sqrt(train['SQBdependency'])
train['edjefe'] = np.sqrt(train['SQBedjefe'])
```

**edjefa:** *years of education of female head of household, based on the interaction of escolari (years of education), head of household and gender, yes=1 and no=0*

So we can just change 'no' to 0 and 'yes' to 1

```python
for col in train.columns:
   train.loc[train['edjefa'] == 'no', col] = 0
   train.loc[train['edjefa'] == 'yes', col] = 1
    
train['edjefa'] = train['edjefa'].astype('int64')
train['edjefa'].head()
```

```python
## Make sure everything worked well
train.dtypes.value_counts()
```

Now to deal with null values.

```python
train.columns[train.isnull().sum() !=0]
```

**v2a1:** Monthly rent payment

**v18q1:** number of tablets household owns

**rez_esc:** Years behind in school

**meaneduc:** Average years of education for adults (18+)

**SQBmeaned:** Square of the mean years of education of adults (>=18) in the household

```python
train[['v2a1', 'v18q1', 'rez_esc', 'meaneduc', 'SQBmeaned']].isnull().sum()
```

For 'meaneduc' and 'SQBmeaned', null values can be replaced with 0 as they represent the mean years of education of adults (>=18) in the household.

```python
train[train['meaneduc'].isnull()][['meaneduc', 'SQBmeaned']]
```

```python
train['meaneduc'] = train['meaneduc'].fillna(0)
train['SQBmeaned'] = train['SQBmeaned'].fillna(0)
```

Since the 'v18q1' column is null if 'v18q' (owns a tablet) is false, we can replace null values with 0.

```python
train[train['v18q1'].isnull()][['v18q1', 'v18q']].head()
```

```python
train['v18q1'] = train['v18q1'].fillna(0)
```

For years behind in school ('rez_esc'), it might be possible to fill null values by examining children in the house since these values probably mean 0 years behind.

```python
train[(train['rez_esc'].isnull() == True) & (train['age'] < 18) & (train['escolari'] > 0)].head()
```

Since the null values are either children or adults that aren't behind in school, we can replace them with 0

```python
train['rez_esc'] = train['rez_esc'].fillna(0)
```

For monthly rent payment 'v2a1', it makes sense to look at if they own the home, rent, homeless, etc.  It looks like null values can be filled with 0 since these people either own the house fully or are in a precarious living situation.

```python
train['v2a1'] = train['v2a1'].fillna(0)
```

```python
# Function to use on test data

def clean_data(data):
    # Dealing with objects in dataframe
    data['dependency'] = np.sqrt(data['SQBdependency'])
    data['edjefe'] = np.sqrt(data['SQBedjefe'])
    for col in data.columns:
        data.loc[data['edjefa'] == 'no', col] = 0
        data.loc[data['edjefa'] == 'yes', col] = 1
    data['edjefa'] = data['edjefa'].astype('int64')
    
    # Dealing with null values
    data['meaneduc'] = data['meaneduc'].fillna(0)
    data['SQBmeaned'] = data['SQBmeaned'].fillna(0)
    data['v18q1'] = data['v18q1'].fillna(0)
    data['rez_esc'] = data['rez_esc'].fillna(0)
    data['v2a1'] = data['v2a1'].fillna(0)
    
    return data
```

```python
test = clean_data(test)
test_set = test.copy()
test_set.set_index('Id', inplace=True)
test_set.drop('idhogar', axis=1, inplace=True)
```

I'll use a LightGBM model to make predictions on the test set.  I've also left in all of the features, but the model could likely be improved by dropping some of them. 

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report
from lightgbm import LGBMClassifier as lgb
```

```python
X = train.drop(['Id', 'idhogar', 'Target'], axis = 1)
y = train['Target']
#X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=101)
```

```python
model = lgb(boosting_type='dart', class_weight='balanced', n_estimators=350, learning_rate=0.3, 
             max_depth=10, importance_type='gain', colsample_bytree=1, random_state=101)
```

```python
kf = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
predicts = []
for train_index, test_index in kf.split(X, y):
    print("###")
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(X_train, y_train, eval_set=[(X_test, y_test)], 
            early_stopping_rounds=20)
    predicts.append(model.predict(test_set))
```

```python
#sorted(zip(model.feature_importances_, train.columns), reverse=True)
```

```python
submission = pd.DataFrame(np.array(predicts).mean(axis=0).round().astype(int),
                             columns=['Target'],
                             index=test_set.index)
```
